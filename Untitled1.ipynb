```python
#申請入學榜單 2022.06.30
#from selenium.webdriver.support.ui import WebDriverWait
#from selenium.webdriver.support import expected_conditions
#from selenium import webdriver
#from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from undetected_chromedriver import Chrome
import time, datetime, random

starttime = datetime.datetime.now() #開始執行時間

browser = Chrome()
url='https://www.com.tw/cross/university_1102_111.html'
browser.get(url)
time.sleep(20)

#先整理系所名單
department_lists=[]
url_lists=[]

for i in range(1):            
    try:
        url_lists.append(browser.find_element(By.XPATH, f'/html/body/div[2]/div[4]/div[2]/table/tbody/tr[{4+i*2}]/td[2]/div[1]/a').get_attribute('href'))
        department_lists.append(browser.find_element(By.XPATH, f'/html/body/div[2]/div[4]/div[2]/table/tbody/tr[{4+i*2}]/td[2]/div[1]/a').text)
    except:
        break
    

for i in range(len(url_lists)):

    #在瀏覽器打上網址連入        
    browser.get(url_lists[i])
    time.sleep(3)

    #for i in range(1):
    for j in range(450): #450

        Yuntech_Admission=''
        STUD_Test_Area=''
        STUD_Name=''
        School_Department=''
        Admission=''
        Admission_Type = ''
        tag=''

        Yuntech_Admission = browser.find_element("xpath",'/html/body/div[2]/div[4]/div[2]/table[1]/tbody/tr['+str(4+j)+']/td[2]/div').text
        print(Yuntech_Admission)
        STUD_Test_Area = browser.find_element("xpath",'/html/body/div[2]/div[4]/div[2]/table[1]/tbody/tr['+str(4+j)+']/td[3]/div/a').text
        print(STUD_Test_Area)
        
        STUD_Name = department_lists[i]+'_學生'+ str(j)
        print(STUD_Name)


endtime = datetime.datetime.now() #執行結束時間
print ('開始時間：'+ str(starttime) +'\n結束時間：'+ str(endtime) +'\n共執行：'+str(endtime - starttime))
```
